```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

* P1.1 Draw square lattice with your own implementation (not build-in
in some library). [1.5P]

```python
def draw_square_lattice(n):
    _, ax = plt.subplots()

    for i in range(n):
        for j in range(n):
            ax.plot(i, j, 'ro')  
            if i > 0:
                ax.plot([i-1, i], [j, j], 'r-')
            if j > 0:
                ax.plot([i, i], [j-1, j], 'r-')

    plt.title(f'{n}x{n} Square Lattice')
    plt.show()

draw_square_lattice(4)
```

* P1.2 Draw chain with your own implementation (not build-in in some
library). [1P]

```python
def draw_chain(n):
    _, ax = plt.subplots()

    for i in range(n):
        ax.plot(i, 0, 'ro')
        if i > 0:
            ax.plot([i-1, i], [0, 0], 'r-')  # Line connecting the points

    plt.title(f'Chain of {n} nodes')
    plt.show()

draw_chain(5)
```

* P1.3 Draw complete graph with your own implementation (not
build-in in some library). [1P]

```python

def draw_complete_graph(n):
    _, ax = plt.subplots()
    ax.set_aspect('equal')
    # Generated by ChatGPT
    angles = np.linspace(0, 2 * np.pi, n, endpoint=False)
    radius = 1 
    points = [(np.cos(angle) * radius, np.sin(angle) * radius) for angle in angles]

    for x, y in points:
        ax.plot(x, y, 'ro')
    
    for i in range(n):
        for j in range(i, n):
            ax.plot([points[i][0], points[j][0]], [points[i][1], points[j][1]], 'r-')

    plt.title(f'Complete Graph with {n} nodes')
    plt.show()

draw_complete_graph(6)
```

* P1.4 Draw weighted graph (any, but N > 5). [1P]

```python
def draw_weighted_graph(n):
    _, ax = plt.subplots()
    ax.set_aspect('equal')

    nodes = [(np.cos(2 * np.pi * i / n), np.sin(2 * np.pi * i / n)) for i in range(n)]

    for x, y in nodes:
        ax.plot(x, y, 'bo')
    
    for i, (x1, y1) in enumerate(nodes):
        for j, (x2, y2) in enumerate(nodes):
            if i < j:
                weight = np.random.randint(1, 10)
                ax.plot([x1, x2], [y1, y2], 'b-')
                # Display the weight in the middle of the edge
                ax.text((x1 + x2) / 2, (y1 + y2) / 2, str(weight), color='black')

    plt.title(f'Weighted Graph with {n} nodes')
    plt.show()

draw_weighted_graph(6)
```

* P1.5 Draw directed graph (any, but N > 5). [1P]

```python
def draw_directed_graph(n):
    connection_probability = 0.3
    random.seed(42)
    _, ax = plt.subplots()
    ax.set_aspect('equal')

    nodes = [(np.cos(2 * np.pi * i / n), np.sin(2 * np.pi * i / n)) for i in range(n)]

    for x, y in nodes:
        ax.plot(x, y, 'ro')

    for i, (x1, y1) in enumerate(nodes):
        for j, (x2, y2) in enumerate(nodes):
            # Generated by ChatGPT
            if i != j and random.random() < connection_probability:
                ax.annotate('', xy=(x2, y2), xytext=(x1, y1),
                            arrowprops=dict(facecolor='red', shrink=0.05))

    plt.title(f'Directed Graph with {n} nodes (Sparse Connections)')
    plt.show()

draw_directed_graph(6)
```
